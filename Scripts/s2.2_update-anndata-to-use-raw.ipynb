```python
import scanpy as sc
import anndata as ad
import numpy as np
import pandas as pd
import pathlib
import os
```

```python
data_dir = pathlib.Path('/home/zhangyc/ShinyServer/FlyCellAtlas-SES/Data')
data_files = [str(item) for item in list(data_dir.rglob("*")) if item.is_file()]
h5ad_files = [f for f in data_files if f.endswith('.h5ad')]
h5ad_files
```

```python
for h5ad_file in h5ad_files:
    print(h5ad_file)
    adata = sc.read_h5ad(h5ad_file)
    h5ad_file_new = os.path.join(os.path.dirname(h5ad_file), os.path.basename(h5ad_file).replace('.h5ad', '_new.h5ad'))
    if adata.raw is not None:
        adata = adata.raw.to_adata()
    adata.write(h5ad_file_new)    
```
